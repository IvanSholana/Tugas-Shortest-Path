```python
import matplotlib.pyplot as plt
import networkx as nx

# Membuat graph
G = nx.Graph()
G.add_nodes_from([1,2,3,4])
G.add_weighted_edges_from([(1,2,0.5), (1,3,1.0), (2,4,2.2), (3,4,1.2)])

# Mendefinisikan posisi node
pos = nx.spring_layout(G)

# Mengambil label node
label = list(G.nodes())

# Mendapatkan koordinat x dan y dari label
x = [pos[i][0] for i in label]
y = [pos[i][1] for i in label]

# Membuat diagram
plt.figure(figsize=(10, 10))

# Menambahkan grid
plt.grid(True)

# Menggambar node dan edge
nx.draw(G, pos, with_labels=True)
labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)

# Menambahkan titik koordinat
plt.scatter(x, y)

# Menambahkan label koordinat
for i, txt in enumerate(label):
    plt.annotate(txt, (x[i], y[i]))

# Memberikan judul pada sumbu x dan y
plt.xlabel('Nilai x')
plt.ylabel('Nilai y')

# Memberikan judul pada diagram
plt.title('Contoh Diagram Kartesius')

# Menampilkan diagram
plt.show()
```

```python
import matplotlib.pyplot as plt

# Membuat data
x = [1,2,3,4,5]
y = [2,4,6,8,10]

# Membuat diagram
plt.figure(figsize=(8, 6)) # menyesuaikan ukuran diagram
plt.plot(x, y)

# Menambahkan label sumbu x dan y
plt.xlabel('Nilai x', fontsize=12) # menambahkan ukuran font
plt.ylabel('Nilai y', fontsize=12) # menambahkan ukuran font

# Menambahkan judul
plt.title('Contoh Diagram Kartesius')

# Menampilkan diagram
plt.show()
```
