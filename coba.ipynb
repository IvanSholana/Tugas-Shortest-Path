```python
import matplotlib.pyplot as plt
import networkx as nx

# Membuat graph
G = nx.Graph()
G.add_nodes_from([1,2,3,4])
G.add_weighted_edges_from([(1,2,0.5), (1,3,1.0), (2,4,2.2), (3,4,1.2)])

# Mendefinisikan posisi node
pos = nx.spring_layout(G)

# Mengambil label node
label = list(G.nodes())

# Mendapatkan koordinat x dan y dari label
x = [pos[i][0] for i in label]
y = [pos[i][1] for i in label]

# Membuat diagram
plt.figure(figsize=(10, 10))

# Menambahkan grid
plt.grid(True)

# Menggambar node dan edge
nx.draw(G, pos, with_labels=True)
labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)

# Menambahkan titik koordinat
plt.scatter(x, y)

# Menambahkan label koordinat
for i, txt in enumerate(label):
    plt.annotate(txt, (x[i], y[i]))

# Memberikan judul pada sumbu x dan y
plt.xlabel('Nilai x')
plt.ylabel('Nilai y')

# Memberikan judul pada diagram
plt.title('Contoh Diagram Kartesius')

# Menampilkan diagram
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        self.label = name + " (" + str(age) + ")"

# Membuat objek-objek Person
p1 = Person("John", 30)
p2 = Person("Jane", 25)
p3 = Person("Bob", 40)

# Membuat graf kosong
G = nx.Graph()

# Menambahkan node ke graf dengan objek Person sebagai label
G.add_node(p1)
G.add_node(p2)
G.add_node(p3)

# Mengubah label pada setiap node
p1.label = "Node 1"
p2.label = "Node 2"
p3.label = "Node 3"

# Menampilkan daftar node dan edge pada graf
print("Nodes: ", G.nodes())
print("Edges: ", G.edges())

# Visualisasi graf dengan label yang telah diubah
nx.draw_networkx(G, with_labels=True, labels=nx.get_node_attributes(G, 'label'))
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

# Membuat objek-objek Person
p1 = Person("John", 30)
p2 = Person("Jane", 25)
p3 = Person("Bob", 40)

# Membuat graf kosong
G = nx.Graph()

# Menambahkan node ke graf dengan objek Person sebagai label
G.add_node(p1)
G.add_node(p2)
G.add_node(p3)

# Menambahkan edge ke graf
G.add_edge(p1, p2)
G.add_edge(p2, p3)
G.add_edge(p3, p1)

# Menampilkan daftar node dan edge pada graf
print("Nodes: ", G.nodes())
print("Edges: ", G.edges())

# Mengubah label pada setiap node
mapping = {p1: "Node 1", p2: "Node 2", p3: "Node 3"}
G = nx.relabel_nodes(G, mapping)

# Menampilkan daftar node dan edge pada graf setelah relabel
print("Nodes: ", G.nodes())
print("Edges: ", G.edges())

# Visualisasi graf dengan label yang telah diubah
nx.draw_networkx(G, with_labels=True)
plt.show()
```

```python
import numpy as np

# Membuat array dengan tipe data Person
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

a = np.array([Person("John", 30), Person("Jane", 25), Person("Bob", 40)])

# Fungsi untuk relabel objek Person menjadi string
def relabel(person):
    return "Name: {}, Age: {}".format(person.name, person.age)

# Menerapkan fungsi relabel pada setiap anggota array dengan np.vectorize
b = np.vectorize(relabel)(a)

# Menampilkan array sebelum dan sesudah relabel
print("Sebelum relabel:")
print(a)
print("Sesudah relabel:")
print(b)
type(b[0])
```

```python
my_dict = {'satu': 1, 'dua': 2, 'tiga': 3, 'empat': 4}

# Mencari key dari value pada dictionary
def get_keys_from_value(dictionary, value):
    return list(filter(lambda x: dictionary[x] == value, dictionary))

# Contoh pemanggilan fungsi get_keys_from_value
keys = get_keys_from_value(my_dict, 2)
print(keys[0])
```

```python
# Deklarasi list
my_list = [True, True, False, True]

# Cek apakah semua elemen dalam list bernilai True
if all(my_list):
    # Jika semua elemen bernilai True, lakukan if statement
    print("Semua elemen dalam list bernilai True")
else:
    # Jika minimal satu elemen bernilai False, lakukan else statement
    print("Minimal satu elemen dalam list bernilai False")
    false_indexes = [i for i, x in enumerate(my_list) if not x]
    print("Indeks dari elemen yang bernilai False: ", false_indexes)
```
